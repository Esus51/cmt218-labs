###Solution

```python
#get last 20 runs
df["Date"] = pd.to_datetime(df["start_date"]).dt.tz_localize(None)
last_20 = (
    df.sort_values("Date", ascending=False)
      .head(20)
      .reset_index(drop=True)
)
iterimlist = [] #initialise a list we can push each set of coords into.
for i, row in last_20.iterrows(): #loop through each row of dataframe
  fld = row["map"] #iterrows returns a dictionary
  # split on single quotes
  parts = fld.split("'")

  # index 7 in Python
  poly = parts[7]

  # turn "\\": into "\" (i.e., unescape)
  poly_clean = poly.encode("utf-8").decode("unicode_escape")
  # Decode
  coords = polyline.decode(poly_clean)  # list of (lat, lon) tuples

  for lat, lon in coords:
    iterimlist.append({
        "Date": row["Date"],
        "lat": lat,
        "lon": lon
    })
runs_location = pd.DataFrame(iterimlist) #put back into dataframe
runs_location.head()
```
