Remember to set appropriate tooltip (hover) text for the chart.

A pie chart is not an `xy` plot. Instead, it is a different type of Plotly chart known as a `domain` plot.

Because of this, pie charts do not use the usual D3-style variables such as: `%{x}`


Instead, you should use the following variables in your hover text:

- `%{label}`   — the category name  
- `%{value}`   — the count for that category  
- `%{percent}` — the percentage of the total

### Appendix

```python
# Bar Chart
fig_distance = px.bar(
    monthly_stats,
    x=monthly_stats["YearMonth"].astype(str),
    y=monthly_stats["TotalDistance"] / 1000,
    title="Total Distance per Month",
    labels={"Month": "Month", "Distance_km": "Distance (km)"}
).update_xaxes(
    tickmode="array",
    tickvals=monthly_stats["YearMonth"].astype(str)[::3],
    tickangle=45
).update_traces(
    marker=dict(color="rebeccapurple"),
    hovertemplate="Month: %{x}<br>Distance: %{y:.2f}"
)

# Scatter Plot
fig_pace = px.scatter(
    Distance_Pace,
    x="Distance_km",
    y="Pace_min_km",
    title="Distance vs Pace",
    labels={"Distance_km": "Distance (km)", "Pace_min_km": "Pace (min/km)"}
).update_traces(
    marker=dict(color="rebeccapurple", opacity=0.5),
    hovertemplate="Distance: %{x:.2f} km<br>Pace: %{y:.2f} min/km"
)

# Line Chart
fig_trailing = px.line(
    trailing365,
    x="Date",
    y="Trailing365",
    title="Trailing 365 Day",
    labels={"Date": "Date", "Trailing365": "Trailing 365 (km)"}
).update_traces(
    line=dict(color="rebeccapurple", width=2)
).update_xaxes(
    dtick="M12",
    tickformat="%Y"
)

# Dashboard
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Distance vs Pace", "Total Distance by Month", "Trailing 365 (km)", "")
)

for t in fig_pace.data:
    fig.add_trace(t, row=1, col=1)

for t in fig_distance.data:
    fig.add_trace(t, row=1, col=2)

for t in fig_trailing.data:
    fig.add_trace(t, row=2, col=1)

# Axis titles
fig.update_xaxes(title_text="Distance (km)", row=1, col=1)
fig.update_yaxes(title_text="Pace (min/km)", row=1, col=1)

fig.update_xaxes(title_text="Month", row=1, col=2)
fig.update_yaxes(title_text="Distance (km)", row=1, col=2)

fig.update_xaxes(title_text="Date", row=2, col=1)
fig.update_yaxes(title_text="Trailing 365 (km)", row=2, col=1)

fig.update_layout(height=900, showlegend=True)
fig.show()
```

```python
#Pie chart solution
fig = px.pie(
    daily_summary,
    values="count",
    names="AMPM",
    title="Runs by Time of Day",
    color="AMPM",
    color_discrete_map={
        "AM": "#E6E6FA",
        "PM": "rebeccapurple"
    }
)

fig.update_traces(
    textposition="inside",
    textinfo="percent+label",
    hovertemplate="%{label}:%{value}"
)

fig.update_layout(
    title_font_size=16,
    legend_title_text="Time of Day"
)

fig.show()
```

```python
# Solution
# Bar Chart
fig_distance = px.bar(
    monthly_stats,
    x=monthly_stats["YearMonth"].astype(str),
    y=monthly_stats["TotalDistance"] / 1000,
    title="Total Distance per Month",
    labels={"Month": "Month", "Distance_km": "Distance (km)"}
).update_xaxes(
    tickmode="array",
    tickvals=monthly_stats["YearMonth"].astype(str)[::3],
    tickangle=45
).update_traces(
    marker=dict(color="rebeccapurple"),
    hovertemplate="Month: %{x}<br>Distance: %{y:.2f}"
)

# Scatter Plot
fig_pace = px.scatter(
    Distance_Pace,
    x="Distance_km",
    y="Pace_min_km",
    title="Distance vs Pace",
    labels={"Distance_km": "Distance (km)", "Pace_min_km": "Pace (min/km)"}
).update_traces(
    marker=dict(color="rebeccapurple", opacity=0.5),
    hovertemplate="Distance: %{x:.2f} km<br>Pace: %{y:.2f} min/km"
)

# Line Chart
fig_trailing = px.line(
    trailing365,
    x="Date",
    y="Trailing365",
    title="Trailing 365 Day",
    labels={"Date": "Date", "Trailing365": "Trailing 365 (km)"}
).update_traces(
    line=dict(color="rebeccapurple", width=2)
).update_xaxes(
    dtick="M12",
    tickformat="%Y"
)

# Pie

fig_ampm = px.pie(
    daily_summary,
    values="count",
    names="AMPM",
    title="Runs by Time of Day",
    color="AMPM",
    color_discrete_map={
        "AM": "#E6E6FA",
        "PM": "rebeccapurple"
    }
).update_traces(
    textposition="inside",
    textinfo="percent+label",
    hovertemplate="%{label}:%{value}"
)

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Distance vs Pace", "Total Distance by Month", "Trailing 365 (km)", "Runs by Time of Day"),
    specs=[
        [{"type": "xy"}, {"type": "xy"}],
        [{"type": "xy"}, {"type": "domain"}]  # pie lives in a domain subplot
    ]
)

for t in fig_pace.data:
    fig.add_trace(t, row=1, col=1)

for t in fig_distance.data:
    fig.add_trace(t, row=1, col=2)

for t in fig_trailing.data:
    fig.add_trace(t, row=2, col=1)

for t in fig_ampm.data:
    fig.add_trace(t, row=2, col=2)

# Axis titles (must be set on the subplot figure)
fig.update_xaxes(title_text="Distance (km)", row=1, col=1)
fig.update_yaxes(title_text="Pace (min/km)", row=1, col=1)

fig.update_xaxes(title_text="Month", row=1, col=2)
fig.update_yaxes(title_text="Distance (km)", row=1, col=2)

fig.update_xaxes(title_text="Date", row=2, col=1)
fig.update_yaxes(title_text="Trailing 365 (km)", row=2, col=1)

fig.update_layout(height=900, showlegend=True)
fig.show()
```
