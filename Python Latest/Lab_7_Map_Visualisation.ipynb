### 2. Add Map Tiles

In Plotly Express, adding map tiles is straightforward—simply switch to `px.line_mapbox()`.

You will also need to specify a `mapbox_style`. In this example, we use `open-street-map`, which is free, open-source, and easy to use. This style provides the background map tiles for the visualisation.

###Solutions

```python
#1
fig = px.line(
    runs_location,
    x="lon",
    y="lat",
    color="Date"
)
fig.show()
```

```python
#2
fig = px.line_mapbox(
    runs_location,
    lat="lat",
    lon="lon",
    color="Date",
    mapbox_style="open-street-map",
    zoom=12
)

fig.update_layout(
    title="Routes by Date"
)

fig.show()
```
