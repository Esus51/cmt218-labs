````python
# Line Chart
fig = px.line(
    trailing365,
    x="Date",
    y="Trailing365",
    title="Trailing 365 Day",
    labels={"Date": "Date", "Trailing365": "Trailing 365 (km)"}
).update_traces(
    line=dict(color="rebeccapurple", width=2)
).update_xaxes(
    dtick="M12",
    tickformat="%Y"
)
fig.show()
````

###Solution

````python
# Ensure Date is datetime
trailing365["Date"] = pd.to_datetime(trailing365["Date"])
trailing365["Year"] = trailing365["Date"].dt.year

fig = px.line(
    trailing365,
    x="Date",
    y="Trailing365",
    title="Trailing 365 Day"
)

fig.update_traces(line=dict(color="rebeccapurple", width=2))

# Build dropdown buttons
years = sorted(trailing365["Year"].unique())

buttons = [
    dict(
        label=str(y),
        method="update",
        args=[
            {"x": [trailing365[trailing365["Year"] == y]["Date"]], #gets the dates for the selected year
             "y": [trailing365[trailing365["Year"] == y]["Trailing365"]]},
            {"title": f"Trailing 365 Day ({y})"}
        ]
    )
    for y in years #Loops through years to create a button for each year
]

fig.update_layout(
    updatemenus=[dict(
        buttons=buttons,
        direction="down",
        showactive=True,
        x=1.05,
        y=1
    )]
)

fig.show()
````
